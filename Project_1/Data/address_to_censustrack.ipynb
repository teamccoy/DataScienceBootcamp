```python
# Import Dependencies
import pandas as pd
import numpy as np
import requests
import json
from pprint import pprint
from config import gkey
```

```python
# Make a reference to the file path
csv_path = "clean_data.csv"

# Read the csv and store into Pandas DataFrame
df = pd.read_csv(csv_path)
df.head(10)
```

```python
listTracts = []
for index, row in df.iterrows():  
    ADDRESS = row["api_address"]
    ADDRESS =str(ADDRESS)
    url = "https://geocoding.geo.census.gov/geocoder/geographies/address?street="+ADDRESS+"&city=San+Diego&state=CA&benchmark=Public_AR_Census2010&vintage=Census2010_Census2010&layers=14&format=json"
    tracts = requests.get(url).json()
    try:
        listTracts.append(tracts['result']['addressMatches'][0]['geographies']['Census Blocks'][0]['TRACT'])
    except:
        listTracts.append(None)

listTracts
```

```python
df['census_tract'] = listTracts
len(df)
```

```python
df.to_csv("even_cleaner_data.csv", encoding="utf-8", index=True, header=True)
```
