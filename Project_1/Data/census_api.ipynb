```python
# Dependencies
import requests
import json
import pandas as pd
from pprint import pprint

# census API Key
from api_keys import api_key

# Starting URL for Census API Call
url = "https://api.census.gov/data/2016/pdb/tract?get=County_name,State_name,Tot_Population_CEN_2010&for=tract:*&in=state:06%20county:073&key=" + api_key 
```

```python
response = requests.get(url).json()
pprint(response)
```

```python
censuspd = pd.DataFrame.from_records(response[1:], columns=response[0])
```

```python
censuspd.to_csv("census_data.csv", encoding="utf-8", index=True, header=True)
```
