```python
# Import Dependencies
import pandas as pd
import gmaps
from config import gkey
import random
import numpy as np

# Access maps with unique API key
gmaps.configure(api_key=gkey)
```

```python
# Make a reference to the books.csv file path
csv_path = "get_it_done_311_requests_datasd_noself.csv"

# Read the csv and store into Pandas DataFrame
# low_memory=False is added to resolve an error that occured when importing the csv
df = pd.read_csv(csv_path, low_memory=False)
df.head(50)
```

```python
# dropping the columns that are not needed
data_df = df[['service_request_id','lat','long','address','description','case_record_type','district']]
data_df.head()
```

```python
# Create a list containing coordinates
# having problems converting "Lat" and "Lng" in this dataframe to float. Using 'lat' and 'long' instead.
data_df = data_df.dropna()


locations = data_df[["lat", "long"]].astype(float)

locations = locations.sample(n=1000)
locations
    
```

```python
# Customize the size of the figure
figure_layout = {
    'width': '1000px',
    'height': '800px',
    'border': '1px solid black',
    'padding': '1px',
    'margin': '0 auto 0 auto'
}
fig = gmaps.figure(layout=figure_layout)
```

```python
# Assign the marker layer to a variable
markers = gmaps.marker_layer(locations)

# Add the layer to the map
fig.add_layer(markers)
fig
```
