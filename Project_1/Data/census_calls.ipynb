```python
# Import Dependencies
import gmaps
import pandas as pd
import requests
import json
from pprint import pprint

# Google developer API key
from config import gkey

# Configure gmaps
gmaps.configure(api_key=gkey)
```

```python
# Make a reference to the books.csv file path
csv_path = "Data/even_cleaner_data.csv"

# Read the csv and store into Pandas DataFrame
df = pd.read_csv(csv_path)
df.head()
```

```python
#setting up the data frame with what is needed
new_df = df[['service_request_id','address','case_record_type','lat','long','sap_problem_type','district','House_Number','Clean_Street','Neighborhood','api_address','census_tract']]
```

```python
#lets' take a look
new_df.head()
```

```python
#reducing the dataframe down to one entry per Neighborhood
hood_df = new_df.drop_duplicates(['Neighborhood'], keep='first')
hood_df = hood_df.set_index('Neighborhood')
hood_df = hood_df.sort_index()
hood_df
```

```python
#preping the groupby for merger into the hood_df
g = new_df.groupby(['Neighborhood'])
g.count()
f = g.count()
f = f.sort_values('address', ascending=False)
f = f.sort_index()
f = f.rename(columns={'address':'count','lat':'count2'})
f = f['count']
f.head()
```

```python
#after aligning the indexes I am concating the two data frames together
heat_df = pd.concat([hood_df,f], axis=1)
heat_df.head()
```

```python
#cutting the heat_df down to just what i need
heat_df = heat_df[['count','service_request_id','lat','long','api_address','case_record_type','census_tract']]
heat_df = heat_df.reset_index()
heat_df.head()
```

```python
heat_df['census_tract'] = heat_df['census_tract'].astype(float)
```

```python
url = "https://api.census.gov/data/2016/pdb/tract?get=County_name,State_name,College_ACS_10_14,Med_HHD_Inc_ACS_10_14,Pop_25_44_CEN_2010&for=tract:*&in=state:06%20county:073&key=cbaa92bbe4b5de472ccaf27bbfbdaed315bbac72"
result = requests.get(url).json()
```

```python
pprint(result)
```

```python
census_df = pd.DataFrame(result)
census_df = census_df.rename(columns=census_df.iloc[0]).drop(census_df.index[0])
census_df['tract'] = census_df['tract'].astype(float)
census_df['census_tract'] = census_df['tract']
census_df = census_df.dropna()
heat_df = heat_df.dropna()
```

```python
census_df.head()
```

```python
merge_df = pd.merge(heat_df,census_df, on =['census_tract'], how = 'inner')
```

```python
merge_df
```

```python
merge_df['25_and_older_with_4yDegree'] = merge_df['College_ACS_10_14']
merge_df['MedianHH_Income'] = merge_df['Med_HHD_Inc_ACS_10_14']
merge_df['Population_25_and_older'] = merge_df['Pop_25_44_CEN_2010']
```

```python
merge_df.to_csv("Neighborhood_census.csv", encoding="utf-8", index=False, header=True)
```
